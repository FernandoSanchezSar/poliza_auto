### 4. Modeling

Con el conjunto de datos casi final, ahora es el momento de crear algunos modelos de regresión diferentes con la Prima como objetivo. Al crear los modelos, exploraremos diferentes parámetros y nos aseguraremos de validar de forma cruzada los hallazgos.

#### 4.1 Seleccionar técnica de modelado

Para predecir el precio de una póliza de automóvil y determinar los principales factores que influyen en este precio, vamos a emplear técnicas de Machine Learning supervisado (regresión) y análisis de importancia de características. Enseguida detallamos el enfoque:

**Regresión lineal Lasso**<br>
Muy interpretativa. Nos permitirá obtener coeficientes asociados a cada característica para identificar los factores más importantes.
Usaremos Lasso para manejar colinealidad o muchas características irrelevantes.

**Random Forest**<br>
Capturan relaciones no lineales entre las características y el precio y permiten obtener la importancia de las características.

**Gradient Boosting XGBoost**<br>
Es un modelo más avanzado y potente que los árboles simples. Ofrecen medidas precisas de importancia de características.
.

#### 4.2 Generar diseño de prueba

Se realizará una división de datos de entrenamiento y pruebas. El 70% de los datos corresponden a entrenamiento y el 30% a pruebas. Esta división lo realizaremos con la función: train_test_split

#### 4.3 Construir modelo

```python
# Preparación de los datos
X = df2.drop(['Prima'], axis = 1)
y = df2['Prima']

# División en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
# Preparar el pipeline
def create_pipeline(model):
    return Pipeline([
        ("poly_scaler", Pipeline([
            ("poly", PolynomialFeatures(degree=3, include_bias=False)),
            ("scaler", StandardScaler())
        ])),
        ("model", model)
    ])

# Definir hiperparámetros para GridSearchCV
param_grid = [
    {
        "model": [Lasso()],
        "model__alpha": [0.01, 0.1, 1, 10]
    },
    {
        "model": [RandomForestRegressor()],
        "model__n_estimators": [50, 100, 200],
        "model__max_depth": [5, 10, None]
    },
    {
        "model": [XGBRegressor()],
        "model__n_estimators": [50, 100, 200],
        "model__max_depth": [3, 5, 7],
        "model__learning_rate": [0.01, 0.1, 0.2]
    }
]

# Configurar GridSearchCV
results = []
for params in param_grid:
    model_name = params["model"][0].__class__.__name__
    pipeline = create_pipeline(params["model"][0])
    grid_search = GridSearchCV(pipeline, param_grid=params, cv=10, scoring="neg_mean_absolute_error", n_jobs=-1)

    # Entrenar y medir el tiempo
    start_time = time.time()
    grid_search.fit(X_train, y_train)
    end_time = time.time()

    # Evaluar en el conjunto de prueba
    best_model = grid_search.best_estimator_
    y_pred = best_model.predict(X_test)
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    # Calcular AUC si aplica (para problemas de clasificación)
    try:
        auc_score = roc_auc_score(y_test, y_pred)
    except ValueError:
        auc_score = None

    # Guardar resultados
    results.append({
        "Modelo": model_name,
        "Mejores parámetros": grid_search.best_params_,
        "MAE": mae,
        "MSE": mse,
        "R²": r2,
        "AUC": auc_score,
        "Tiempo de entrenamiento (s)": round(end_time - start_time, 2)
    })

# Convertir resultados a DataFrame
df_results = pd.DataFrame(results)
print(df_results)
```

```python
# Segunda etapa: Comparar coeficientes y Permutation Importance con nombres de características
comparison_results = []
for params in param_grid:
    model_name = params["model"][0].__class__.__name__
    pipeline = create_pipeline(params["model"][0])
    grid_search = GridSearchCV(pipeline, param_grid=params, cv=10, scoring="neg_mean_absolute_error", n_jobs=-1)
    grid_search.fit(X_train, y_train)
    best_model = grid_search.best_estimator_

    # Obtener nombres de las características generadas por PolynomialFeatures
    poly_features = best_model.named_steps["poly_scaler"].named_steps["poly"].get_feature_names_out(input_features=X.columns)

    # Obtener coeficientes e intercepción (si aplica)
    if hasattr(best_model.named_steps["model"], "coef_"):
        coefficients = pd.DataFrame({
            "Feature": poly_features,
            "Coefficient": best_model.named_steps["model"].coef_
        }).sort_values(by="Coefficient", ascending=False)
        intercept = best_model.named_steps["model"].intercept_
    else:
        coefficients = None
        intercept = None

    # Permutation Importance
    perm_importance = permutation_importance(best_model, X_test, y_test, n_repeats=10, random_state=42)
    perm_importance_df = pd.DataFrame({
        "Feature": X.columns,
        "Importance": perm_importance.importances_mean
    }).sort_values(by="Importance", ascending=False)

    # Guardar resultados
    comparison_results.append({
        "Modelo": model_name,
        "Coeficientes": coefficients,
        "Intercepción": intercept,
        "Permutation Importance": perm_importance_df
    })

# Convertir comparación a un DataFrame o guardar como lista de resultados detallados
for result in comparison_results:
    print(f"Modelo: {result['Modelo']}")
    if result["Coeficientes"] is not None:
        print("\nCoeficientes:")
        print(result["Coeficientes"].head(10))  # Mostrar solo los 10 coeficientes más significativos
    print("\nPermutation Importance:")
    print(result["Permutation Importance"].head(10))  # Mostrar solo las 10 características más importantes
    print("-" * 50)
```
